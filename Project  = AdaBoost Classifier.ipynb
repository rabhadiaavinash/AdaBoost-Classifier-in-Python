```python
import pandas as pd
```

```python
dataset = pd.read_csv('c:\Users\Kishore\Desktop\Datasets\mushrooms.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset = dataset.sample(frac = 1)
```

```python
dataset.head()
```

```python
dataset.columns= ['class','cap-shape','cap-surface','cap-color','bruises','odor','gill-attachment','gill-spacing','gill-size','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring','stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population','habitat'
]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
dataset.columns
```

```python
for label in dataset.columns:
    dataset[label] = LabelEncoder().fit_transform(dataset[label])
```

```python
print(dataset.info())
```

```python
dataset.head()
```

```python
X = dataset.drop(['class'] , axis = 1)
Y = dataset['class']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X ,Y ,test_size= 0.30 , random_state = 40 )
```

```python
x_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion = 'entropy' , max_depth =1)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
AdaBoost = AdaBoostClassifier(base_estimator = model , n_estimators =400 , learning_rate =1)
```

```python
boostmodel = AdaBoost.fit(x_train,y_train)
```

```python
y_pred = boostmodel.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```
